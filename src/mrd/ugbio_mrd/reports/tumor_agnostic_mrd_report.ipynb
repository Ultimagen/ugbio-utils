# Tumor-Agnostic MRD Analysis Report

This report presents the results of a tumor-agnostic MRD (Minimal Residual Disease) analysis.
The analysis examines trinucleotide substitution profiles across VAF (Variant Allele Frequency) bins
and applies COSMIC mutational signature deconvolution using MuSiCal.

```python
import os

import IPython.display
import pandas as pd
from IPython.display import HTML, Image
```

```python
# input parameters (injected by papermill)
trinuc_histogram_png = None
trinuc_counts_csv = None
signature_plot_png = None
signature_weights_csv = None
basename = None
```

## Trinucleotide Substitution Profile by VAF

The following histogram shows the trinucleotide substitution profile split by VAF bins.
Each panel represents a VAF range, showing the distribution of the 96 canonical trinucleotide changes.

```python
if trinuc_histogram_png and os.path.isfile(trinuc_histogram_png):
    IPython.display.display(Image(filename=trinuc_histogram_png))
else:
    IPython.display.display(HTML('<p style="color:red;">Trinucleotide histogram not available.</p>'))
```

### Trinucleotide Counts Table

Counts of each trinucleotide substitution type per VAF bin.

```python
if trinuc_counts_csv and os.path.isfile(trinuc_counts_csv):
    df_counts = pd.read_csv(trinuc_counts_csv)
    IPython.display.display(df_counts)
else:
    IPython.display.display(HTML('<p style="color:red;">Trinucleotide counts CSV not available.</p>'))
```

## COSMIC Signature Deconvolution

MuSiCal was used to decompose the trinucleotide profiles into COSMIC mutational signatures.
The barplots below show the relative contribution of each signature in each VAF bin.

```python
if signature_plot_png and os.path.isfile(signature_plot_png):
    IPython.display.display(Image(filename=signature_plot_png))
else:
    IPython.display.display(HTML('<p style="color:red;">Signature deconvolution plot not available.</p>'))
```

### Signature Weights Table

The estimated contribution (weight) of each COSMIC signature per VAF bin.

```python
if signature_weights_csv and os.path.isfile(signature_weights_csv):
    df_weights = pd.read_csv(signature_weights_csv, index_col=0)
    IPython.display.display(df_weights)
else:
    IPython.display.display(HTML('<p style="color:red;">Signature weights CSV not available.</p>'))
```
